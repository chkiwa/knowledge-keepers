```python
import pandas as pd
df = pd.read_excel('filtered_data.xlsx')
```

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("facebook/nllb-200-distilled-600M")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")
df["Problem_Ar"] = ""
for i, problem in enumerate(df['Problem']):
    english_tokens = tokenizer(problem, return_tensors="pt")
    arabic_tokens = model.generate(**english_tokens, forced_bos_token_id=tokenizer.lang_code_to_id["arb_Arab"], max_length=400)
    arabic_text = tokenizer.batch_decode(arabic_tokens, skip_special_tokens=True)[0]
    df.loc[i, "Problem_Ar"] = arabic_text  
    
    # Print the original and translated problems
    print("Original Problem: ", problem)
    print("Translated Problem: ", arabic_text)

```
